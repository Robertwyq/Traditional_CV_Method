```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('./img/1.jpg')

copyIma = img1.copy()
h, w = img1.shape[:2]
print(h, w)
mask = np.zeros([h+2, w+2], np.uint8)

plt.subplot(2, 1, 1)
plt.imshow(img1, cmap='gray')

retval, image, mask, rect = cv2.floodFill(img1, mask, (100, 100),
                                          (0, 0, 0), (4, 4, 4), (20, 20, 20))

plt.subplot(2, 1, 2)
plt.imshow(image, cmap='gray')
plt.show()
```
